```python
from selenium import webdriver
from selenium.webdriver.common.by import By
from selenium.webdriver.common.keys import Keys 
from selenium.webdriver.support.ui import WebDriverWait
from selenium.webdriver.support import expected_conditions as EC
import csv
import time 

# Launch the Chrome browser
driver = webdriver.Chrome()
driver.get("https://www.amazon.com/")

# Wait until the search box is clickable
for_search = WebDriverWait(driver, 10).until(
    EC.element_to_be_clickable((By.ID, "twotabsearchtextbox"))
)
for_search.clear()
for_search.send_keys("keyboards")
for_search.send_keys(Keys.RETURN)

# Wait for the search results container
for_div = WebDriverWait(driver, 10).until(
    EC.presence_of_element_located((By.CLASS_NAME, "puisg-col-inner"))
)

# Try to find the heading element
h1 = WebDriverWait(driver, 10).until(
    EC.presence_of_all_elements_located((By.CSS_SELECTOR, '[data-cy="title-recipe"]'))
)

product_link = WebDriverWait(driver,10).until(
    EC.presence_of_all_elements_located((By.CLASS_NAME,"a-link-normal"))
)

product_img = WebDriverWait(driver,10).until(
    EC.presence_of_all_elements_located((By.CSS_SELECTOR, '[data-cy="image-container"]'))
)

# url = product_img.find_element(By.CLASS_NAME,"s-image")


reviews = WebDriverWait(driver,10).until(
    EC.presence_of_all_elements_located((By.CSS_SELECTOR, '[data-cy="reviews-block"]'))
)

price_recipe = WebDriverWait(driver,10).until(
    EC.presence_of_all_elements_located((By.CSS_SELECTOR, '[data-cy="price-recipe"]'))
)

price_recipe = WebDriverWait(driver,10).until(
    EC.presence_of_all_elements_located((By.CSS_SELECTOR, '[data-cy="price-recipe"]'))
)

delivery_recipe = WebDriverWait(driver,10).until(
    EC.presence_of_all_elements_located((By.CSS_SELECTOR, '[data-cy="delivery-recipe"]'))
)

data = []

for i in range(min(len(h1), len(product_link), len(product_img), len(reviews), len(price_recipe), len(delivery_recipe))):

    product_data = {
        "Product": h1[i].text,
        "Link": product_link[i].get_attribute('href'),
        "Image": product_img[i].find_element(By.CLASS_NAME, "s-image").get_attribute('src'),
        "Reviews": reviews[i].text,
        "Price": price_recipe[i].text,
        "Delivery": delivery_recipe[i].text
    }

    data.append(product_data)

    print(f"""
Product {i+1}:
Product_Title: {product_data["Product"]}
product_link: {product_data["Link"]}
product_img = {product_data["Image"]}
Reviews: {product_data["Reviews"]}
Price: {product_data["Price"]}
Delivery: {product_data["Delivery"]}
    """)


driver.quit()


with open("products.csv", "w", newline="", encoding="utf-8") as file:
    writer = csv.writer(file)
    writer.writerow(["Product", "Link", "Image", "Reviews", "Price", "Delivery"])

    for product in data:
        writer.writerow([product["Product"], product["Link"], product["Image"], product["Reviews"], product["Price"], product["Delivery"]])

print("✅ CSV file saved successfully!")
```
